```python
import numpy as np
import pandas as pd
import math
```

```python
df = pd.read_csv('numbats.csv')



```

```python
df
```

```python

df['decade'] = df['year'].apply(lambda x: math.floor(x / 10) * 10 if not pd.isnull(x) else np.nan)
df['fiveyear'] = pd.cut(df['year'], bins=np.arange(1850, 2025, 5))
df['eventDate'] = pd.to_datetime(df['eventDate'], errors="coerce")
df['quarter'] = df['eventDate'].dt.quarter.fillna('NaN')
cases = [[df['quarter'] == 1, df['quarter'] == 2, df['quarter'] == 3, df['quarter'] == 4], ["summer", "fall", "winter", "spring"]]
df['season'] = np.select(cases[0], cases[1], 'undefined')

```
